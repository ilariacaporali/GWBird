```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_ETd, strain_ETd = np.loadtxt('psd/et_d.txt', unpack=True)
f_coba , strain_coba = np.loadtxt('psd/ET_Sh_coba.txt', unpack=True)
```

```python
# plot

plt.figure(figsize=(10, 6))
plt.plot(f_ETd, strain_ETd, label='ET_d')
plt.plot(f_coba, strain_coba, label='ET_Sh_coba')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Strain [1/sqrt(Hz)]')
plt.grid(axis='both', alpha= 0.2)
plt.legend()
plt.show()
```
